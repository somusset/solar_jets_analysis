# Extract useful information from catalogue

What I need is: for each SOL event, the FOV, the time range, and the time ranges in which we have jets (and possibly the position of these jets)

In the jet catalogue, several jets (also called jet clusters) can be coming from the same SOL event. A SOL event can be seen as a data set by itself: part of it contain (a least) a jet, and part of it does not contain a jet. Since SOL events can be repeated in the list of jets, we reduce the list of SOL events so that each SOL event appears only once

```python
unique_SOL_events = np.unique(Cluster_SOL)
```

```python
len(unique_SOL_events)
```

We will query the HEK database to find information about these SOL events but for this we need to know what kind of time range we are talking about

```python
print(np.min(Cluster_date))
print(np.max(Cluster_date))
```

Apparently we are exploring data from 2011 to 2016 (good, that's expected)

Now query the HEK data base for CJ events (Coronal jets) during this period of time. Since we query on several years it will take a few seconds

```python
from sunpy.net import attrs as a, Fido
timerange = a.Time('2011/01/01 00:00:00', '2017/01/01 00:00:00')
res = Fido.search(timerange, a.hek.CJ) 
```

### Example with the first event from the SJH catalogue

Let's see if we can find events matching our first SOL event from the Jet Hunter list

```python
res['hek'][res['hek']["SOL_standard"]==unique_SOL_events[0]]
```

We actually have two rows that share the same "unique" SOL id, that's annoying, but almost all the fields are identical in the two rows, as we can see below

```python
selection0 = res['hek'][res['hek']["SOL_standard"]==unique_SOL_events[0]]
```

```python
print(selection0['event_starttime'])
print(selection0['event_endtime'])
```

```python
fov0 = selection0['hpc_bbox'] ## stands for helioprojective coordinates of the bounding box
fov0
```

Since the two rows have identical values in the fields we are interested in, we will always keep only the data from the first row

Extraction of the field of view information into 4 float variables

```python
x1 = float(fov0.value[0].strip('POLYGON()').split(',')[0].split()[0])
x2 = float(fov0.value[0].strip('POLYGON()').split(',')[1].split()[0])
y1 = float(fov0.value[0].strip('POLYGON()').split(',')[0].split()[1])
y2 = float(fov0.value[0].strip('POLYGON()').split(',')[2].split()[1])
x1
```

# Request the data (as cutouts)

Now we will download the images corresponding to this time and fov using this example
https://docs.sunpy.org/en/stable/generated/gallery/acquiring_data/downloading_cutouts.html#sphx-glr-generated-gallery-acquiring-data-downloading-cutouts-py

```python
import os

import matplotlib.pyplot as plt

import astropy.units as u
from astropy.coordinates import SkyCoord
from astropy.time import Time
from astropy.visualization import ImageNormalize, SqrtStretch

import sunpy.coordinates  # NOQA
import sunpy.map
from sunpy.net import Fido
from sunpy.net import attrs as a
```

```python
start_time = selection0['event_starttime'][0]
end_time = selection0['event_endtime'][0]
bottom_left = SkyCoord(x1*u.arcsec, y1*u.arcsec, obstime=start_time, observer="earth", frame="helioprojective")
top_right = SkyCoord(x2*u.arcsec, y2*u.arcsec, obstime=start_time, observer="earth", frame="helioprojective")
print(bottom_left)
print(top_right)
```

```python
cutout = a.jsoc.Cutout(bottom_left, top_right=top_right, tracking=False)
```

```python
jsoc_email = "sophie.musset@esa.int"
```

Now we look for files. We input the start and end time, the wavelength (304 A), we want data taken at a cadence of 24 seconds

```python
query = Fido.search(
    a.Time(start_time, end_time),
    a.Wavelength(304*u.angstrom),
    a.Sample(24*u.s),
    a.jsoc.Series.aia_lev1_euv_12s,
    a.jsoc.Notify(jsoc_email),
    a.jsoc.Segment.image,
    cutout,
)
print(query)
```

Only now we download the files. It will takes a while (could be a few minutes)

```python
files = Fido.fetch(query, overwrite=True)
files.sort()
```

```python
print(files[0])
```

Now load in map sequence (useful to plot)

```python
sequence = sunpy.map.Map(files, sequence=True)
```

Plot the first map in the sequence, without any axes

```python
smap = sequence[0]

figure = plt.figure(frameon=False)
ax = plt.axes([0, 0, 1, 1])
# Disable the axis
ax.set_axis_off()

# Plot the map.
# Since we are not interested in the exact map coordinates,
# we can simply use :meth:`~matplotlib.Axes.imshow`.
norm = smap.plot_settings['norm']
norm.vmin, norm.vmax = np.percentile(smap.data, [1, 99.9])
ax.imshow(smap.data,
          norm=norm,
          cmap=smap.plot_settings['cmap'],
          origin="lower")
plt.show()
```

Plot the first map in the sequence as we would usually do

```python
smap = sequence[0]


fig = plt.figure()
ax = fig.add_subplot(projection=smap)
smap.plot(axes=ax)
plt.colorbar()

plt.show()
```

## Which part(s) of the event contain(s) jet(s)?

Now let's create a flag that says if the data in the map sequence countains a jet or not

```python
contain_jet = np.zeros(len(sequence), dtype=bool) # for the moment initialized with only '0' everywhere
```

I also want an array that contains all the dates from the different maps in the map sequence. There is probably a easy way to extract this but I do not know how so I do it ulgy:

```python
dates = []
for map in sequence:
    dates.append(map.date)
```

Find the jets from the SJH catalogue that belongs to this SOL event

```python
sjh_identifier[Cluster_SOL == unique_SOL_events[0]]
```

In this case there is only one jet in this SOL event, but in other cases there will be several, so we will loop anyway

```python
jets = sjh_identifier[Cluster_SOL == unique_SOL_events[0]]
for jet in jets:
    start_jet = Cluster_date[sjh_identifier == jet][0]
    duration_jet = stat_dur[sjh_identifier == jet][0] ## in minutes
    end_jet = start_jet + np.timedelta64(int(duration_jet*60),'s')
    print(start_jet)
    print(end_jet)
    contain_jet = ((dates >= start_jet)&(dates <= end_jet))|contain_jet
print(contain_jet)
```
